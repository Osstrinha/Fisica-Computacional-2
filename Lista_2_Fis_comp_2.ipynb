```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
from scipy.optimize import minimize_scalar





print("A implementao abaixo foi implementada mediante uso do Gemini e GPT, com pequenas alteraes")
print()




# Constantes fsicas
G = 6.67430e-11  # Constante gravitacional universal (m/kg/s)
M_sol = 1.989e30  # Massa do Sol (kg)
M_halley = 2.2e14  # Massa aproximada do cometa Halley (kg)
velocidade_afelio = 6.12e3  # Velocidade no aflio (m/s)
distancia_afelio = 5.28e12  # Distncia no aflio (m)

# Mtodo de Picard para integrar a equao de movimento
def picard(pos, vel, M, dt):
    acc = gravidade(pos, M)  # Acelerao inicial
    pos_new = pos + vel * dt + 0.5 * acc * dt**2  # Nova posio
    acc_new = gravidade(pos_new, M)  # Nova acelerao
    vel_new = vel + 0.5 * (acc + acc_new) * dt  # Nova velocidade
    return pos_new, vel_new

# Funo para calcular a acelerao gravitacional
def gravidade(pos, M):
    r = np.linalg.norm(pos)  # Distncia ao Sol
    return -G * M * pos / r**3

# Simulao da rbita
def simular_orbita(massa_estrela, massa_cometa, distancia_afelio, velocidade_afelio, dt, tmax):
    # Condies iniciais
    pos = np.array([distancia_afelio, 0])  # Posio inicial no aflio
    vel = np.array([0, velocidade_afelio])  # Velocidade inicial no aflio
    M = massa_estrela

    # Listas para armazenar os resultados
    posicoes = [pos]
    velocidades = [vel]
    tempos = [0]
    perihelio_dist = np.linalg.norm(pos)  # Inicializa a menor distncia ao Sol
    vel_max = np.linalg.norm(vel)  # Inicializa a velocidade mxima

    # Loop de simulao
    t = 0
    while t < tmax:
        pos, vel = picard(pos, vel, M, dt)
        posicoes.append(pos)
        velocidades.append(vel)
        tempos.append(t)

        # Atualiza a menor distncia ao Sol (perilio)
        perihelio_dist = min(perihelio_dist, np.linalg.norm(pos))
        # Atualiza a velocidade mxima
        vel_max = max(vel_max, np.linalg.norm(vel))

        # Verifica se completou uma rbita (retornou ao aflio)
        if t > 0 and np.allclose(pos, [distancia_afelio, 0], atol=1e9) and np.dot(vel, [0, 1]) > 0:
            break

        t += dt

    # Converte os resultados para arrays
    posicoes = np.array(posicoes)
    velocidades = np.array(velocidades)
    tempos = np.array(tempos)

    return posicoes, velocidades, tempos, perihelio_dist, t / (60 * 60 * 24 * 365.25), vel_max

# Parmetros de simulao
dt = 1e5  # Passo de tempo (s)
tmax = 2.5e9  # Tempo mximo de simulao (s)

# Simulao
posicoes, velocidades, tempos, perihelio_dist, periodo_orbita, vel_max = simular_orbita(
    M_sol, M_halley, distancia_afelio, velocidade_afelio, dt, tmax
)

# Resultados
print(f"(a) Distncia mnima ao Sol (perilio): {perihelio_dist:.2e} m")
print(f"(b) Tempo necessrio para uma rbita completa: {periodo_orbita:.2f} anos")

print("c)")
# Grfico da rbita
plt.figure(figsize=(8, 8))
plt.plot(posicoes[:, 0], posicoes[:, 1], label="rbita do cometa")
plt.scatter([0], [0], color="orange", label="Sol (centro de massa)", s=200)
plt.xlabel("x (m)")
plt.ylabel("y (m)")
plt.title("rbita do cometa Halley ao redor do Sol")
plt.legend()
plt.axis("equal")
plt.grid(True)
plt.show()



print(f"(d) Mdulo da velocidade mxima: {vel_max:.2e} m/s")


print("6)")

# Given constants
omega_0 = 0.77  # Linear mass density (kg/m)
T0_initial = 150  # Initial tension (N)
f_A = 5  # Height at point A (m)
dfdx_A = -0.1445  # Initial slope at point A
x_A = 0  # Starting point (horizontal position of A)
distance_AB = 25  # Horizontal distance between points A and B

# Differential equation system
def catenary_eq(x, y, T0):
    # y[0] = f(x), y[1] = f'(x)
    dfdx = y[1]
    d2fdx2 = (omega_0 / T0) * np.sqrt(1 + dfdx**2)
    return [dfdx, d2fdx2]

# (a) Solving for the height of point B (distance_AB = 25m)
def solve_catenary(T0, x_span, y0, num_points=100):
    x_values = np.linspace(x_span[0], x_span[1], num_points)
    solution = solve_ivp(catenary_eq, x_span, y0, args=(T0,), t_eval=x_values, method='RK45')
    return solution.t, solution.y[0]

# Initial conditions
y0 = [f_A, dfdx_A]

# Solve for part (a) using T0 = 150 N
x_values, f_values = solve_catenary(T0_initial, [x_A, x_A + distance_AB], y0)

# (a) Output height at point B
f_B = f_values[-1]
print("Height of point B:", f_B, "m")

# Plot the shape of the catenary curve for visualization
plt.figure(figsize=(8, 6))
plt.plot(x_values, f_values, label="Catenary Curve")
plt.scatter([x_A, x_A + distance_AB], [f_A, f_B], color='red', label="Points A and B")
plt.xlabel("Horizontal distance x (m)")
plt.ylabel("Vertical distance f(x) (m)")
plt.legend()
plt.title("Catenary Curve of the Rope")
plt.grid()
plt.show()


# Differential equation system
def catenary_eq(x, y, T0):
    # y[0] = f(x), y[1] = f'(x)
    dfdx = y[1]
    d2fdx2 = (omega_0 / T0) * np.sqrt(1 + dfdx**2)
    return [dfdx, d2fdx2]

# Solving the catenary equation
def solve_catenary(T0, x_span, y0, num_points=100):
    x_values = np.linspace(x_span[0], x_span[1], num_points)
    solution = solve_ivp(catenary_eq, x_span, y0, args=(T0,), t_eval=x_values, method='RK45')
    return solution.t, solution.y[0]

# (a) Calculate height at point B with initial conditions
y0 = [f_A, dfdx_A]
x_values, f_values_a = solve_catenary(T0_initial, [x_A, x_A + distance_AB], y0)
f_B_a = f_values_a[-1]
print("Part (a) - Height at point B:", f_B_a, "m")

# (b) Adjust slope f'(x=A) to make height at point B equal to f_A
def objective_slope(dfdx_A_new):
    y0_new = [f_A, dfdx_A_new]
    _, f_values = solve_catenary(T0_initial, [x_A, x_A + distance_AB], y0_new)
    return abs(f_values[-1] - f_A)

result_slope = minimize_scalar(objective_slope)
dfdx_A_b = result_slope.x
y0_b = [f_A, dfdx_A_b]
x_values, f_values_b = solve_catenary(T0_initial, [x_A, x_A + distance_AB], y0_b)
print("Part (b) - New slope f'(x=A) for equal heights:", dfdx_A_b)
print("Part (b) - Height at point B with adjusted slope:", f_values_b[-1], "m")

# (c) Adjust tension T0 to make height at point B equal to f_A
def objective_tension(T0_new):
    _, f_values = solve_catenary(T0_new, [x_A, x_A + distance_AB], y0)
    return abs(f_values[-1] - f_A)

result_tension = minimize_scalar(objective_tension, bounds=(50, 300), method='bounded')
T0_c = result_tension.x
x_values, f_values_c = solve_catenary(T0_c, [x_A, x_A + distance_AB], y0)
print("Part (c) - New tension T0 for equal heights:", T0_c, "N")
print("Part (c) - Height at point B with adjusted tension:", f_values_c[-1], "m")

# (d) Find minimum distance to the ground for each solution
f_min_a = np.min(f_values_a)
f_min_b = np.min(f_values_b)
f_min_c = np.min(f_values_c)
print("Part (d) - Minimum height for part (a):", f_min_a, "m")
print("Part (d) - Minimum height for part (b):", f_min_b, "m")
print("Part (d) - Minimum height for part (c):", f_min_c, "m")

# Plotting all three solutions for visualization
plt.figure(figsize=(10, 8))
plt.plot(x_values, f_values_a, label="Part (a) - Initial Condition")
plt.plot(x_values, f_values_b, label="Part (b) - Adjusted Slope")
plt.plot(x_values, f_values_c, label="Part (c) - Adjusted Tension")
plt.scatter([x_A, x_A + distance_AB], [f_A, f_B_a], color='red', label="Points A and B")
plt.xlabel("Horizontal distance x (m)")
plt.ylabel("Vertical distance f(x) (m)")
plt.legend()
plt.title("Catenary Curves for Different Conditions")
plt.grid()
plt.show()




print()
print("7)")
print()


# Constants
k = 1           # Force constant
m = 1           # Mass
dt = 0.0001     # Time step
t_max = 20      # Maximum simulation time
r0 = np.array([1.0, 0.0])  # Initial position vector (in the x-direction)
v0 = np.array([0.5, 1.0])  # Initial velocity vector

# Simulation function using Euler method
def simulate_orbit(alpha):
    # Initial conditions
    r = r0.copy()  # Position vector
    v = v0.copy()  # Velocity vector

    # Time steps
    num_steps = int(t_max / dt)
    positions = np.zeros((num_steps, 2))  # Store positions for plotting

    # Euler method for integration
    for i in range(num_steps):
        # Save current position
        positions[i] = r

        # Compute distance and force
        r_magnitude = np.linalg.norm(r)
        if r_magnitude == 0:
            break  # Prevent division by zero

        # Force calculation
        force_magnitude = -k * r_magnitude ** alpha
        force = force_magnitude * r / r_magnitude  # Central force in radial direction

        # Acceleration
        a = force / m

        # Update velocity and position
        v += a * dt
        r += v * dt

    return positions

# Values of alpha to test
alphas = [2, 1, 0, -1, -2, -3]
plt.figure(figsize=(12, 8))

for alpha in alphas:
    positions = simulate_orbit(alpha)
    plt.plot(positions[:, 0], positions[:, 1], label=f" = {alpha}")

# Plot settings
plt.xlabel("x")
plt.ylabel("y")
plt.title("Trajectories for Different  Values")
plt.legend()
plt.axis('equal')
plt.grid()
plt.show()
```
